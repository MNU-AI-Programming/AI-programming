`week04_datafiles/grade_tests.csv`ë¥¼ ì½ì–´ì„œ, ë‚´ ì½”ë“œê°€ ê¸°ëŒ€ê°’ê³¼ ë§ëŠ”ì§€ í™•ì¸í•˜ëŠ” ì—°ìŠµì…ë‹ˆë‹¤.


````python
import os, csv

test_path = "week04_datafiles/grade_tests.csv"
if not os.path.exists(test_path):
    print("í…ŒìŠ¤íŠ¸ íŒŒì¼ì´ ì—†ìŠµë‹ˆë‹¤. zipì„ í’€ì—ˆëŠ”ì§€ í™•ì¸í•˜ì„¸ìš”.")
else:
    tests = []
    with open(test_path, "r", encoding="utf-8") as f:
        reader = csv.DictReader(f)
        for row in reader:
            tests.append((int(row["score"]), row["expected"]))
    print("í…ŒìŠ¤íŠ¸ ê°œìˆ˜:", len(tests))
    print("ì˜ˆì‹œ:", tests[:3])

    # TODO: ì•„ë˜ í•¨ìˆ˜ì— ë³¸ì¸ ë“±ê¸‰ ë¡œì§ì„ ì˜®ê²¨ ë„£ê³  í…ŒìŠ¤íŠ¸í•˜ì„¸ìš”.
    def grade_of(score: int) -> str:
        if score < 0 or score > 100:
            return "ERROR"
        if score >= 90:
            return "A"
        elif score >= 80:
            return "B"
        elif score >= 70:
            return "C"
        elif score >= 60:
            return "D"
        else:
            return "F"

    ok = 0
    for sc, exp in tests:
        got = grade_of(sc)
        if got == exp:
            ok += 1
        else:
            print("âŒ ë¶ˆì¼ì¹˜:", sc, "expected=", exp, "got=", got)

    print(f"âœ… í†µê³¼: {ok}/{len(tests)}")
````
