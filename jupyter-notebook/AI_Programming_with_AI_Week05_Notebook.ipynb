## 6ï¸âƒ£ ì‹¤ìŠµ 1: ì„±ì  ë“±ê¸‰ í•¨ìˆ˜ ë¶„í•´


```python
def get_grade(score: int) -> str:
    if score >= 90: return "A"
    elif score >= 80: return "B"
    elif score >= 70: return "C"
    elif score >= 60: return "D"
    else: return "F"

assert get_grade(95) == "A"
assert get_grade(50) == "F"
```
